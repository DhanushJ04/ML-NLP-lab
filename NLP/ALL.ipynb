# 1.  Regular Expressions

```python
# Date format
```

```python
import re

text = "my dob is 20/10/2004"
date_pattern = r"\b(?:0[1-9]|[12][0-9]|3[0-1])/(?:0[1-9]|1[0-2])/\d{4}\b"
dates = re.findall(date_pattern, text)
print("Dates found : ", dates)
```

```python
# Password matching
```

```python
import re

password_pattern = r"^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*?&])[A-Za-z\d@%!%?&]{8,}$"
password = "Abc@1234"
if re.match(password_pattern, password):
    print("Password is valid")
else:
    print("Password is invalid")
```

```python
# Sentence ending with ?
```

```python
import re 

txt = """What is NLP?
NLP is a branch of AI.
How does regex work?
Python is easy.
Why should we learn NLP?"""

question_pattern = r"[^?]*\?"
question = [line for line in txt.splitlines()
            if re.fullmatch(question_pattern, line)]
print("Question found : ", question)
```

```python
# Extract hexadecimal color code
```

```python
import re

text = "Colors used #FF5733, #ab, #123ABC, #333"
hex_pattern = r"#(?:[A-Fa-f0-9]{6}|[A-Fa-f0-9]{3})\b"
colors = re.findall(hex_pattern, text)
print("Hex color codes found : ", colors)
```

# 3. POS Tagging and Stop Word removal

```python
import spacy

nlp = spacy.load("en_core_web_sm")
sentence = "SpaCy is a powerful library for Natural Language Processing."
doc = nlp(sentence)

# POS Tagging
print("POS Tagging : ")
for token in doc:
    print(token.text, "->", token.pos_)

# Remove stopwords
filwords = [token.text for token in doc if not token.is_stop and not token.is_punct]
filsent = " ".join(filwords)
print("\nFiltered Sentence : ", filsent)
```

# 4. Lemmatization and Stemming

```python
import spacy
from nltk.stem import PorterStemmer
import nltk

nlp = spacy.load("en_core_web_sm")
stemmer = PorterStemmer()
sentence = "The boys are running and studies are ongoing."
doc = nlp(sentence)

# Stemming
words = sentence.split()
print("Stemming : ")
for word in words:
    print(word, "->", stemmer.stem(word))

# Lemmatization
print("\nLemmatization : ")
for token in doc:
    print(token.text, "->", token.lemma_)
```

# 6. CFG

```python
import nltk
from nltk import CFG
from nltk.parse import ChartParser

# Define the grammar
grammar = CFG.fromstring("""
S -> NP VP
NP -> Det N
VP -> V NP
Det -> 'The'|'an'
N -> 'boy'|'apple'
V -> 'eats'
""")

# Create Parser
parser = ChartParser(grammar)

# Sentence to parse
sentence = "The boy eats an apple"
words = sentence.split()

# Parse the sentence
for tree in parser.parse(words):
    print(tree)
    tree.pretty_print()
```

# 7. Lexical Relations

```python
import spacy
from nltk.corpus import wordnet
import nltk

nlp = spacy.load("en_core_web_sm")
sentence = "The dog chased the cat in the garden."
doc = nlp(sentence)

# Extract nouns
nouns = [token.text for token in doc if token.pos_ == "NOUN"]
print("Noun : ", nouns)

# Hypernyms
print("\nHypernyms : ")
for noun in nouns:
    synsets = wordnet.synsets(noun)
    if synsets:
        hypernyms = synsets[0].hypernyms()
        if hypernyms:
            print(noun, "-?", hypernyms[0].lemma_names())
```

# 10. Top-Down parsing

```python
import nltk
from nltk import CFG
from nltk.parse import RecursiveDescentParser

# Define the grammar
grammar = CFG.fromstring("""
S -> NP VP
NP -> Det N
VP -> V NP
Det -> 'The'|'the'
N -> 'boy'|'ball'
V -> 'kicked'
""")

# sentence to parse
sentence = "The boy kicked the ball"
words = sentence.split()

# Top-Down parser
parser = RecursiveDescentParser(grammar)
print("Top-Down Parsing : \n")
for tree in parser.parse(words):
    print(tree)
    tree.pretty_print()
```

# 11. Bottom-Up Parsing

```python
import nltk
from nltk import CFG
from nltk.parse import ShiftReduceParser

# Define the grammar
grammar = CFG.fromstring("""
S -> NP VP
NP -> Det N
VP -> V NP
Det -> 'The'|'the'
N -> 'boy'|'ball'
V -> 'kicked'
""")

# sentence to parse 
sentence = "The boy kicked the ball"
words = sentence.split()

# Bottom-Up parser
parser = ShiftReduceParser(grammar)
print("Bottom-Up Parser : \n")
for tree in parser.parse(words):
    print(tree)
    tree.pretty_print()
```

# 12. Chatbot

```python
import nltk
from nltk.tokenize import word_tokenize

print("Welcome to AIMIT Admission Chatbot")
print("Type 'exit' to end the conversation.\n")

# Knowledge base
responses = {
    "courses": "AIMIT offers programs like MCA, MBA, MSc Data Science and other postgraduate courses.",
    "eligibility": "For MCA admission, candidates must have a Bachelor's degree with Mathematics or Statistics.",
    "fees": "The course fees varies by program. Please visit the official AIMIT website or contact the admission office.",
    "admission": "Admission are based on academic performance and university guidelines.",
    "contact": "You can contact AIMIT at 0824-2286881/0824-2286882 or visit the admission office.",
    "location": "AIMIT is located near Beeri, Kotekar, Mangalore, Karnataka.",
    "hostel": "Hostel facilities are available for both boys and girls with good amenities."
}

# Chatbot loop
while True:
    user_input = input("You : ").lower()
    if user_input == "exit":
        print("Chatbot : Thank you for your interest in AIMIT. Goodbye!")
        break

    tokens = word_tokenize(user_input)

    if "course" in tokens or "program" in tokens:
        print("Chatbot : ", responses["courses"])
    elif "eligibility" in tokens or "qualifications" in tokens:
        print("Chatbot : ", responses["eligibility"])
    elif "fee" in tokens or "fees" in tokens:
        print("Chatbot : ", responses["fees"])
    elif "admission" in tokens or "apply" in tokens:
        print("Chatbot : ", responses["admission"])
    elif "contact" in tokens or "phone" in tokens:
        print("Chatbot : ", responses["contact"])
    elif "location" in tokens or "phone" in tokens:
        print("Chatbot : ", responses["location"])
    elif "hostel" in tokens:
        print("Chatbot : ", responses["hostel"])
    else:
        print("Chatbot : Sorry, I didn't understand. Please ask about courses, admission, fees, hostel or contact details")
```

# 14. Word Frequency

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text = """
Natural Language Processing is a branch of Artificial Intelligence.
It helps computers understand human language.
Many application such as chatbots and sentiment analysis use NLP.
"""

# Tokenization
words = word_tokenize(text)

# stop words 
stop_words = set(stopwords.words("english"))

# word frequency dictionary
word_freq = {}
for word in words:
    word = word.lower()
    if word not in stop_words and word.isalpha():
        if word not in word_freq:
            word_freq[word] = 1
        else:
            word_freq[word] += 1

# Display
print("Word Frequency : \n")
for word, freq in word_freq.items():
    print(word, ":", freq)
```
